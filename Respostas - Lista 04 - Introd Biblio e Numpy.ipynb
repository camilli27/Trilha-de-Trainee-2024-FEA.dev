A proposta desse desafio  justamente testar essa hiptese ! 

Temos uma sequncia de arranjos que correspondem a "camadas" do caf e precisamos ir afunilando e remodelando para que possamos ir somando at formar um nico valor (que corresponde ao nosso "caf filtrado"). 

##### PS: alguns gros esto acima do valor mximo que a cafeteira suporte. Substitua gros maiores que 50 (gramas) pelo valor 30 

```python
# Camadas de gros de caf para serem filtrados. Para fins didticos imagine que so gros inteiros e isso no afeta o processo
# Camadas de gros de caf para serem filtrados. Para fins didticos imagine que so gros inteiros e isso no afeta o processo
camada_1 = np.array([[5, 2, 7],
                     [12, 5, 9]])

camada_2 = np.array([[15, 8],
                     [4, 10],
                     [52, 65]])

camada_3 = np.array([[17, 11, 19],
                     [6, 22, 26]])

camada_4 = np.array([[78, 15],
                     [16, 23],
                     [12, 32]])
```

Tendo essas camadas em vista, crie um cdigo que remodele (reshape/resize) os arranjos para uma dimenso comum e possibilite a soma. Ao final printe a soma de todos os valores. 

_Lembre-se de corrigir gros maiores que 50 para um valor 30 (seria uma margem de erro para a mquina)_

```python
# Seu cdigo comea aqui

camadas = [camada_1,camada_2,camada_3,camada_4]
tamanho_padrao = (3,2)

for i,camada in enumerate(camadas):
    camadas[i] = np.reshape(camada,tamanho_padrao)
    camadas[i][camadas[i] > 50] = 30

soma_total = np.sum(camadas, axis=None)
print(soma_total)
```
